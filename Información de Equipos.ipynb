```python
#Importar las bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Obtener la informaci√≥n
full_data = pd.read_csv('./data/OraclesElixir/FullData.csv')
```

```python
teams = np.unique(full_data['team'])
teams = np.reshape(teams,(192,1))
```
